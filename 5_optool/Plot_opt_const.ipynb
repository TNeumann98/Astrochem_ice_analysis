### Plotting routine for optical depth calculated with optool
testing on 100CO_26CO2_10K

with the following standard setups:

d = 4e-05

n0 = 1.22

n2 = 1.73

error = 0.001

```python
%matplotlib notebook
#read in data from ym script
import numpy as np
import matplotlib.pyplot as plt
from mymisc import *

path1 = './results/dustkappa_d0.00025_n01.16_n21.73_err0.0001_pureCO2_10K_quick5.inp'
path2 = './results/dustkappa_d0.00025_n01.22_n21.73_err0.0001_100CO_4CO2_10K_quick5_2.inp'
path3 = './results/dustkappa_d0.00025_n01.22_n21.73_err0.0001_100CO_26CO2_10K_quick5_2.inp'
path4 = './results/dustkappa_d0.00025_n01.22_n21.73_err0.0001_100CO_70CO2_10K_quick5_2.inp'
path5 = './results/dustkappa_d5e-05_n01.16_n21.73_err0.0001_100H2O_10CO2_10K_quick5.inp'

lam, k_tn, n_tn = np.loadtxt(path1, delimiter=' ',usecols=(3,6,9), comments = '#', skiprows=28, unpack=True)
lam2, k_tn2, n_tn2 = np.loadtxt(path2, delimiter=' ',usecols=(3,6,9), comments = '#', skiprows=28, unpack=True)
lam3, k_tn3, n_tn3 = np.loadtxt(path3, delimiter=' ',usecols=(3,6,9), comments = '#', skiprows=28, unpack=True)
lam4, k_tn4, n_tn4 = np.loadtxt(path4, delimiter=' ',usecols=(3,6,9), comments = '#', skiprows=28, unpack=True)
lam5, k_tn5, n_tn5 = np.loadtxt(path5, delimiter=' ',usecols=(3,6,9), comments = '#', skiprows=28, unpack=True)
```

```python
path3 = './lnk_data_old/dustkappa_co-a-Palumbo2006.txt'
lamp, k_tnp, n_tnp = np.loadtxt(path3, delimiter=' ', usecols=(3,6,9), comments = '#', unpack=True)
```

```python
# plt.plot(1e4/lam_wr,k_wr,label = 'Original code')
plt.plot(lam, k_tn, "-", label="pureCO2_10K absorption", alpha=0.6)
plt.plot(lam3, k_tn3, ":", label="100CO_26CO2_10K absorption")
plt.plot(lam4, k_tn4, ":", label="100CO_70CO2_10K absorption")
plt.plot(lam2, k_tn2, ":", label="100CO_4CO2_10K absorption")
plt.plot(lam5, k_tn5, "-.", label="100H2O_10CO2_10K absorption", alpha=0.6)  # , alpha = 0.5)
#plt.plot(lamg, k_tng, "--", label="Gerakines2020 absorption", alpha=0.8)
plt.plot(lamp, k_tnp, "--", label="Palumbo2006 absorption")
plt.xlabel("wavelength [$\mu$m]")
plt.ylabel(r"opacity [$cm^2/g$]")
plt.legend()
plt.yscale("log")
plt.xscale("log")

plt.xlim(12,20)
plt.ylim(3e2,1e4)
#plt.title('k-value check')
plt.savefig('./plots/comparison_absorption_zoom.pdf', dpi='figure', format='pdf')
```
